## Nested Lists (2D Lists) — Quick Scaffolding

A **nested list** (also called a **2D list**) is a list where each item is itself a list.
We often use nested lists to represent **grids** like seating charts, game boards, or tables.

### 1) A small example (2 rows × 3 columns)

```python
grid = [
    ["EMPTY", "EMPTY", "EMPTY"],
    ["EMPTY", "EMPTY", "EMPTY"]
]
```
### 2) Building a grid with loops (recommended)

To build a grid, we usually make rows one at a time and add them to the chart.
```
def make_chart(rows, cols):
    chart = []
    for r in range(rows):              # outer loop = rows
        row = []
        for c in range(cols):          # inner loop = columns
            row.append("EMPTY")
        chart.append(row)
    return chart
```

### 3) Printing a grid neatly (nested loops)

```
def print_chart(chart):
    for r in range(len(chart)):
        for c in range(len(chart[0])):
            print(chart[r][c], end=" ")
        print()  # new line after each row
```

### 4) Updating one cell
```
chart[2][3] = "Ava"   # row 2, col 3
```

### 5) Common reminders

- The outer loop usually goes through the rows

- The inner loop usually goes through the columns

- Use len(chart) for number of rows

- Use len(chart[0]) for number of columns (assuming all rows are same length)

### F3. Mini Gradebook (nested list) (2D list + loops + functions)
Create a mini gradebook for students and assignments using a 2D list.

**Requirements:**
- Build a 5×4 grid (5 students, 4 assignments) filled with `0`
- Write a function `set_score(gradebook, student, assignment, score)` that updates a score
- Write a function `print_gradebook(gradebook)` that prints each student’s scores neatly
- randomize grades selecting 4 in a range of 65-100  
- Use nested loops to compute and return tuple:
  - each student’s **average score**
  - the **overall class average**
  - Find the student (row) with the highest total score.
  - print returned tuple using an f-string




````python
# F3 answer here

# creating empty gradebook
import random
gradebook = []
for i in range(5):
  row = []
  for j in range(4):
    row.append(0)
  gradebook.append(row)

# function to set the score at a specific index
def set_score(gradebook, student, assignment, score):
  gradebook[student][assignment] = score
  return gradebook

# looping through the list to set the grades using set_score
for i, row in enumerate(gradebook):
  for j in range(len(row)):
    set_score(gradebook, i, j, random.randint(65, 100))

# printing the gradebook
def print_gradebook(gradebook):
  for row in gradebook:
    print(row)

print(print_gradebook(gradebook))

def calculations(gradebook):
  # finding each students average
  averages = []
  total_sum = 0
  total_count = 0
  highest_score = 0
  highest_score_row_index = 0
  for i, row in enumerate(gradebook):
    row_sum = 0
    row_count = 0
    for j in range(len(row)):
      row_sum += row[j]
      row_count += 1
      total_sum += row[j]
      total_count += 1
      if (i==0 and j==0) or gradebook[i][j] > highest_score:
        highest_score = gradebook[i][j]
        highest_score_row_index = i
    averages.append(row_sum/row_count)
  total_avg = total_sum / total_count

  return averages, total_avg, highest_score, highest_score_row_index

averages, total_avg, highest_score, highest_score_row_index = calculations(gradebook)

# prints information
for i, avg, in enumerate(averages):
  print(f"Student {i+1} average: {avg}")
print(f"Class average: {total_avg}")
print(f"The highest score was {highest_score} which was acheived by Student {highest_score_row_index+1}")

# final f string
print(f"The student averages are {averages}, the class average is {total_avg}, the highest score was {highest_score}, and it was acheived by the student at row index {highest_score_row_index}")
````
